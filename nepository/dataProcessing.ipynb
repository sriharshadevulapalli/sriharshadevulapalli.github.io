```python
group_dict = {}
counter = 0
for ip in unique_people:
    if len(df.loc[df['Person'] == ip, 'Family'].values) >0:
        breakout_ip = df.loc[df['Person'] == ip, 'Family'].values[0]
        if breakout_ip not in group_dict:
            counter += 1
            group_dict[breakout_ip] = counter
        else:
            pass
```

```python
temp_links_list = list(grouped_src_dst.apply(lambda row: {"source": row['source'], "target": row['target'], "value": row['count']}, axis=1))
```

```python
links_list = []
for link in temp_links_list:
    record = {"value":link['value'], "source":unique_people.get_loc(link['source']),
     "target": unique_people.get_loc(link['target'])}
    links_list.append(record)
```

```python

nodes_list = []

for ip in unique_people:
    if len(df.loc[df['Person'] == ip, 'Family'].values) >0:
        breakout_ip = df.loc[df['Person'] == ip, 'Family'].values[0]
    nodes_list.append({"name":ip, "group": group_dict.get(breakout_ip)})
```

```python
group_dict
```

```python
group_list = []
for link in group_dict:
    record = {"key":link, "value":group_dict[link]}
    group_list.append(record)
```

```python
json_prep = {"nodes":nodes_list, "links":links_list,"groupKey":group_list}

json_prep.keys()
```

```python
json_dump = json.dumps(json_prep, indent=1, sort_keys=True)
```
